# line graph

```python
import pandas as pd
import numpy as np

dates = pd.date_range(start = '2022-01-01', periods=20)
print(dates)

np.random.seed(123)
stockprice = np.round(np.random.uniform(1477.20,1602.95,20),2)
print(stockprice)
```

```python
plt.figure(figsize=(15,3), dpi = 300)

plt.plot(dates, stockprice, color='green')

plt.title("Line chart to show Stock price for 30 days", fontsize = 15, color = 'b')
plt.xlabel("Dates", fontsize = 15, color = 'black')
plt.ylabel("Stock price", fontsize = 15, color = 'black')

plt.xticks(list(dates), rotation = 45) 
plt.yticks(range(1400,1600,25)) 

plt.show()
```
