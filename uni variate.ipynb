### plot.bar

```python
sales['Region'].value_counts().plot.bar()
```

```python
sns.countplot(x='Region',data=sales,)
```

```python
plt.figure(figsize=(15,4))
plt.title('number of order of the region')
sns.countplot(x='Region',data=sales)
```

```python
colour = sns.color_palette("cubehelix")
plt.figure(figsize=(10,5))
sns.countplot(x='Region', data=sales, palette=colour)
plt.title('Number of Orders by Region')
plt.ylabel('Number of Orders')
plt.xlabel('Unit_MakingCost')
plt.xticks(rotation=90)
plt.show
```

```python
colour = sns.color_palette("cubehelix")
plt.figure(figsize=(10,5))
sns.countplot(x='Total_Revenue', data=sales, palette=colour)
plt.title('Number of Orders by Region')
plt.ylabel('Number of Orders')
plt.xlabel('Region')
plt.xticks(rotation=90)
plt.show
```

# pie chart

```python
sales['Region'].value_counts().unique()
```

```python
sales['Region'].value_counts().plot.pie()
```

```python
counts = sales['Region'].value_counts()
plt.pie(counts,labels=counts.unique(),autopct='%.0f%%')
plt.show
```

```python
counts = sales['Region'].value_counts()
plt.pie(counts,labels=counts.index,autopct='%.0f%%')
plt.show
```

```python
colors = sns.color_palette("Paired")
plt.pie(sales['Region'].value_counts(), autopct='%.0f%%',
        explode=[0,0,0,0,0,0,0], colors=sns.color_palette("Blues"))
plt.show()
```

```python
colors = sns.color_palette('flare')
plt.pie(sales['Total_Revenue'].value_counts(),autopct= '%.0f%%',
      colors = colors)
plt.show
```
